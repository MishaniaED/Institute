```python
polar_plot(sqrt, 0, 2 * pi, fill=True)
```

```python
html("<h2 style='color: orange'>Заливка между двумя графиками</h1>")
```

```python
polar_plot(cos(4*x) + 1.5, 0, 2*pi, fill=0.5 * cos(4*x) + 2.5, fillcolor='orange')
```

```python
html("<h2 style='color: orange'>Заливка между несколькими графиками</h1>")
```

```python
polar_plot([(1.2+k*0.2)*log(x) for k in range(6)], 1, 3 * pi, fill={0: [1], 2: [3], 4: [5]})
```

```python
html("<h2 style='color: orange'>Исключение точек на разрывах</h1>")
```

```python
polar_plot(log(floor(x)), (x, 1, 4*pi), exclude=[1..12])
```
