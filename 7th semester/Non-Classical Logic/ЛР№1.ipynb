Ядро множества D состоит из элементво интервала (0, 14) 

Выберем элемент 8

µ_A(8) = 0.5

µ_B(8) = 0.5

µ_C(8) = 0

µ_!C(8) = 1 - µ_C(8) = 1

µ_(A(8) ∩ B(8)) = µ_A(8) * µ_B(8) = 0,5 * 0,5 = 0,25

µ_(A ∩ B ∪ !C) = µ_A(8) * µ_B(8) + µ_!C(8) - µ_A(8) * µ_B(8) * µ_!C(8) = 0,25 + 1 - 0,25 * 1 = 1
