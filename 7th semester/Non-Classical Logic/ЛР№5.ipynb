**Аксиомы**

$
Электронные\;устройства - ЭУ, Персональные\;компьютеры - ПК\\
(1) ПК = Ноутбук \cup Стационарный\;компьютер\\
(2) Ноутбук \cap Стационарный\;компьютер = \emptyset\\
(3) Игровой \subseteq Ноутбук\\
(4) ЭУ = Носимые \cup Стационарные\\
(5) Носимые \cap Стационарные = \emptyset\\
(6) ПК \subseteq ЭУ\\
(7) Грфический \subseteq Стационарный\;компьютер\\
(8) \neg Серверные = Грфический \cup Игровой\\
$

**Утверждения**

$
  (a) Ноутбук \cup Стационарный\;компьютер = ЭУ\\
  (b) Носимые \cup Стационарные = ПК\\
  (c) Игровой \subseteq ПК\\
  (d) Графический \subseteq ПК\\
  (e) ПК \subseteq Носимые \cup Стационарные\\\
  (f) \negСерверные \subseteq ПК\\
  (g) \negНосимые \cap \negСтационарные = \negЭУ\\
$

**Проверка истинности утверждений**

$
a)\space Из \space(1) \spaceи \space(6): Ноутбук \cup Стационарный\;компьютер \subseteq ЭУ.\space Неверно, \spaceт. к. \space ПК \ne ЭУ\\
b)\space Из \space(6)\space и \space(7): Носимый \ne Ноутбук, \space Стационарный \ne Стационарный \space компьютер \space (в \space подмножестве \space  могут \space  быть \space  иные \space  элементы) => нельзя \space применить\space (1). \space Неверно.\\
c)\space Игровой \subseteq Ноутбук \subseteq ПК. \space Верно.\\
d)\space Графический \subseteq Стационарный компьютер \subseteq ПК.\space Верно.\\
e)\space ПК \subseteq ЭУ = Носимые \cup Стационарные. \spaceВерно.\\
f)\space \neg Серверные \subseteq Стационарный \space компьютер \cup Ноутбук = ПК. \spaceВерно.\\
g)\space Применяя \space закон \space де \space Моргана \space к \space (4) \space получим \space (g). \space Верно.\\
