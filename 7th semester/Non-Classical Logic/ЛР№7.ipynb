```python
g = rdflib.Graph()
g.parse("ЛР№6.owl")
```

```python
query1 = """
 PREFIX rdf: <http://www.w3.org/1999/02/22-rdf-syntax-ns#>
PREFIX owl: <http://www.w3.org/2002/07/owl#>
PREFIX rdfs: <http://www.w3.org/2000/01/rdf-schema#>
PREFIX xsd: <http://www.w3.org/2001/XMLSchema#>
PREFIX a: <http://www.semanticweb.org/misha/ontologies/2022/11/untitled-ontology-2#>
  SELECT ?name ?exp ?edu
    WHERE {
          ?prog rdf:type a:TYPE.
          ?prog a:Имя ?name.
          ?prog a:Опыт_работы ?exp.
          ?prog a:Образование ?edu.
    }
"""

list_type = ['Программист', 'Мастер_по_натсройке', 'Мастер_по_ремонту']
list_fild = ['Name', 'Experience', 'Education']

for j in list_type:
  print(f'Type: {j}')
  vals = []
  for i in g.query(query1.replace('TYPE', j) ):
    vals.append(i)
  print(tabulate.tabulate(vals, list_fild, tablefmt="grid") + '\n')
```
